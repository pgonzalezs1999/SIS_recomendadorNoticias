## ElPais Tecnologia

```python
html = requests.get("https://elpais.com/tecnologia/")
soup = BeautifulSoup(html.content, 'html.parser')

listaNoticias = soup.findAll("article")

linkNoticias = list()

for a in listaNoticias:
        link = "https://elpais.com" + a.find("a")["href"]
        linkNoticias.append(link)
    
print(linkNoticias)
```

```python
titulos = list()
entradillas = list()
textos = list()
parrafos = list()
fechas = list()
nNoticia = 0
superArray = list()
tags = []
textTag = list()

for i in linkNoticias[0:50]:
    
    try:
        
        htmlText = requests.get(i)
        soup = BeautifulSoup(htmlText.content, 'html.parser')
        
        titulo = soup.find(class_="a_t").text
        titulos.append(titulo)
        
        entradilla = soup.find(class_="a_st").text
        entradillas.append(entradilla)
        
        divCentral = soup.find("div", class_="a_c")
        parrafos = divCentral.findAll("p", class_="")
        texto = ""
        for p in parrafos:
            texto += p.text + " "
        textos.append(texto)
        
        # tags
        tags_html = soup.find(class_="_df _ls").findAll("li")        
        
        tags = []
        for tag in tags_html:
            tag = tag.find("a").text
            textTag.append(tag)
            
            tags.append(tag)
        
        superArray.append(tags)
        
        textTag.append("##") 
                        
        fecha = soup.find(id="article_date_p")["data-date"]
        fechas.append(fecha)

        dfNoticia["Noticias"] = dfNoticia["Noticias"].append("Noticia" + nNoticia)

        nNoticia += 1
    except:
        continue
```

```python
for i in superArray:
    print(i)
```

```python
textos[0] = textos[0].strip() # Elimina los espacios del comienzo y fin de la cadena
#textos[0] = textos[0][1:]  #Esta linea elimina el primer caracter
for i in textos:
    print(i)
```

```python
dfElPaisTecnologia = pd.DataFrame(list(zip(titulos, entradillas, textos, fechas, superArray)), columns = ['Titulos','Entradillas', 'Textos', 'Fechas', 'Etiquetas'])
dfElPaisTecnologia = dfElPaisTecnologia.assign(Noticiario= "ElPais")
dfElPaisTecnologia = dfElPaisTecnologia.assign(Categoria= "Tecnologia")
dfElPaisTecnologia
```

```python
dfElPaisTecnologia = dfElPaisTecnologia.reindex(columns=['Titulos','Entradillas', 'Textos', 'Fechas', 'Noticiario', 'Categoria', 'Etiquetas'])
```

```python
dfElPaisTecnologia
```

```python
dfElPaisTecnologia["Fechas"] = dfElPaisTecnologia["Fechas"].replace(to_replace="[A-Z][0-9]{2}:\d{2}:\d{2}.\d{3}[A-Z]|[A-Z][0-9]{2}:\d{2}:\d{2}[A-Z]", value = ' ', regex = True)
dfElPaisTecnologia
```

```python
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```
