```python
from tkinter import *
import pandas as pd
```

```python
def CrearVentana():
    nuevoRoot = Tk()
    nuevoRoot.title("Grupo X - Comparacion de noticias") # Escribir nombre de la ventana
    nuevoRoot.resizable(False, False) # Si se puede escalar a lo ancho y a lo alto respectivamente
    nuevoRoot.iconbitmap("icono256.ico") # Escoger imagen que acompaña al titulo
    return nuevoRoot
```

```python
def CrearFrame(nuevoColor):
    nuevoFrame = Frame(root) # Asociar a la ventana
    nuevoFrame.config(bg=nuevoColor) #Establecer un color al contenedor   
    # Esta linea siempre tiene que ir al final de la funcion:
    nuevoFrame.pack() # Reescalar el frame cuando se reescale la ventana
    return nuevoFrame
```

```python
def ElegirCarpeta():
    rutasElegidas = list()
    df = pd.read_csv("rutas.csv")
    for i in range(len(df)):
        if(valorMostradoMedio.get().replace(" ", "") in df["rutas"][i] and valorMostradoCategoria.get() in df["rutas"][i]):
            rutasElegidas.append(df["rutas"][i])
    valorMostradoNoticias.set(rutasElegidas[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
    desplegableNoticias = OptionMenu(frameSeleccion, valorMostradoNoticias, *rutasElegidas) # Crear el desplegable
    desplegableNoticias.grid(row=4, column=1, padx=(0, 20), sticky="w") # Ubicar el desplegable
```

```python
def BuscarNoticias():
    if(valorMostradoNoticias.get() != "No hay noticias"):
        f = open(valorMostradoNoticias.get(), "r")
        leer = f.read()
        contenedorPreview = Text(frameSeleccion, width=60, height=14, padx=10) # Crear el cuadro de texto de varias filas
        contenedorPreview.insert(END, leer)
        scrollPreview = Scrollbar(frameSeleccion, command=contenedorPreview.yview) # Crear y escalar el scroll, y asociarlo a su cuadro de texto
        scrollPreview.grid(row=1, column=3, rowspan=6, padx=(0, 20), pady=(0, 20), sticky="nsw") # Ubicar el scroll
        contenedorPreview.grid(row=1, column=2, rowspan=6, padx=(20, 0), pady=(0, 20)) # Ubicar el cuadro de texto de varias filas
        contenedorPreview.config(yscrollcommand=scrollPreview.set) # Para que el scroll muestre a qué altura se encuentra el cuadro de texto asociado
        f.close()
```

```python
root = CrearVentana()  # Crear frame de la ventana
frameSeleccion = CrearFrame("#EEEEEE") # Crear contenedor para un grupo de elementos (como un div de HTML)
frameResultados = CrearFrame("#EEEEEE")

labelSeleccion = Label(frameSeleccion, text="Seleccionar noticia de referencia:").grid(row=0, column=0, columnspan=4, sticky="w", padx=30, pady=(20, 20)) # Crear y ubicar el texto

labelMedioSeleccion = Label(frameSeleccion, text="Medio:").grid(row=1, column=0, sticky="w", padx=(30, 20)) # Crear y ubicar el texto
opcionesMedio = ["El Mundo", "El Pais", "20 minutos"] # Opciones que se mostrarán en el desplegable del medio
valorMostradoMedio = StringVar(frameSeleccion) # Para recoger el dato seleccionado
valorMostradoMedio.set(opcionesMedio[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableMedio = OptionMenu(frameSeleccion, valorMostradoMedio, *opcionesMedio) # Crear el desplegable
desplegableMedio.grid(row=1, column=1, padx=(0, 20), sticky="w") # Ubicar el desplegable

labelCategoriaSeleccion = Label(frameSeleccion, text="Categoría:").grid(row=2, column=0, sticky="w", padx=(30, 20)) # Crear y ubicar el texto
opcionesCategoria = ["Salud", "Tecnologia", "Ciencia"] # Opciones que se mostrarán en el desplegable de la categoría
valorMostradoCategoria = StringVar(frameSeleccion) # Para recoger el dato seleccionado
valorMostradoCategoria.set(opcionesCategoria[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableCategoria = OptionMenu(frameSeleccion, valorMostradoCategoria, *opcionesCategoria) # Crear el desplegable
desplegableCategoria.grid(row=2, column=1, padx=(0, 20), sticky="w") # Ubicar el desplegable

botonAplicar = Button(frameSeleccion, text="Aplicar", command=ElegirCarpeta).grid(row=3, column=0, padx=(30, 20), pady=(10, 10), columnspan=2, sticky="ew") # Crear y ubicar el botón

labelNoticiaSeleccion = Label(frameSeleccion, text="Noticias:").grid(row=4, column=0, sticky="w", padx=(30, 20)) # Crear y ubicar el texto
opcionesNoticias = ["No hay noticias"] # Opciones que se mostrarán en el desplegable de las noticias
valorMostradoNoticias = StringVar(frameSeleccion) # Para recoger el dato seleccionado
valorMostradoNoticias.set(opcionesNoticias[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableNoticias = OptionMenu(frameSeleccion, valorMostradoNoticias, *opcionesNoticias) # Crear el desplegable
desplegableNoticias.grid(row=4, column=1, padx=(0, 20), sticky="w") # Ubicar el desplegable

labelTOPNSeleccion = Label(frameSeleccion, text="TOP-N:").grid(row=5, column=0, sticky="w", padx=(30, 20)) # Crear y ubicar el texto
opcionesTOPN = [2, 3, 4, 5, 6, 7, 8, 9, 10] # Opciones que se mostrarán en el desplegable del TOP-N
valorMostradoTOPN = StringVar(frameSeleccion) # Para recoger el dato seleccionado
valorMostradoTOPN.set(opcionesTOPN[3]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableTOPN = OptionMenu(frameSeleccion, valorMostradoTOPN, *opcionesTOPN) # Crear el desplegable
desplegableTOPN.grid(row=5, column=1, padx=(0, 20), sticky="w") # Ubicar el desplegable

botonBuscar = Button(frameSeleccion, text="Buscar", command=BuscarNoticias).grid(row=6, column=0, padx=(30, 20), pady=(10, 10), columnspan=2, sticky="ew") # Crear y ubicar el botón

labelPreview = Label(frameSeleccion, text="Preview:").grid(row=0, column=2, sticky="sw", padx=20, pady=(20, 0)) # Crear y ubicar el texto
contenedorPreview = Text(frameSeleccion, width=60, height=14, padx=10) # Crear el cuadro de texto de varias filas
scrollPreview = Scrollbar(frameSeleccion, command=contenedorPreview.yview) # Crear y escalar el scroll, y asociarlo a su cuadro de texto
scrollPreview.grid(row=1, column=3, rowspan=6, padx=(0, 20), pady=(0, 20), sticky="nsw") # Ubicar el scroll
contenedorPreview.grid(row=1, column=2, rowspan=6, padx=(20, 0), pady=(0, 20)) # Ubicar el cuadro de texto de varias filas
contenedorPreview.config(yscrollcommand=scrollPreview.set) # Para que el scroll muestre a qué altura se encuentra el cuadro de texto asociado

labelFiltrar = Label(frameResultados, text="Filtrar:").grid(row=0, column=0, padx=(20, 0), pady=(20, 20), sticky="e") # Crear y ubicar el texto
opcionesFiltrar = ["Todos", "El Mundo", "El Pais", "20 minutos"] # Opciones que se mostrarán en el desplegable de Filtrar
valorMostradoFiltrar = StringVar(frameResultados) # Para recoger el dato seleccionado
valorMostradoFiltrar.set(opcionesFiltrar[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableFiltrar = OptionMenu(frameResultados, valorMostradoFiltrar, *opcionesFiltrar) # Crear el desplegable
desplegableFiltrar.grid(row=0, column=1, padx=(0, 20), pady=(20, 20), sticky="w") # Ubicar el desplegable

labelRanking = Label(frameResultados, text="Ranking:").grid(row=0, column=2, padx=(20, 0), pady=(20, 20), sticky="e") # Crear y ubicar el texto
opcionesRanking = ["No hay noticias"] # Opciones que se mostrarán en el desplegable del ranking
valorMostradoRanking = StringVar(frameResultados) # Para recoger el dato seleccionado
valorMostradoRanking.set(opcionesRanking[0]) # Por defecto se mostrará y se escogerá la opción indicada en el enunciado
desplegableRanking = OptionMenu(frameResultados, valorMostradoRanking, *opcionesRanking) # Crear el desplegable
desplegableRanking.grid(row=0, column=3, padx=(0, 20), pady=(20, 20), sticky="w") # Ubicar el desplegable

botonVerRecom = Button(frameResultados, text="Ver recomendacion", command=VerRecom).grid(row=0, column=4, padx=(30, 20), pady=(10, 10), columnspan=2, sticky="ew") # Crear y ubicar el botón

labelNoticia = Label(frameResultados, text="Texto de la noticia recomendada:").grid(row=1, column=0, columnspan=6, sticky="sw", padx=20) # Crear y ubicar el texto
contenedorNoticia = Text(frameResultados, width=80, height=14, padx=10) # Crear el cuadro de texto de varias filas
scrollNoticia = Scrollbar(frameResultados, command=contenedorNoticia.yview) # Crear y escalar el scroll, y asociarlo a su cuadro de texto
scrollNoticia.grid(row=2, column=6, padx=(0, 20), pady=(0, 20), sticky="nsw") # Ubicar el scroll
contenedorNoticia.grid(row=2, column=0, columnspan=6, padx=(20, 0), pady=(0, 20)) # Ubicar el cuadro de texto de varias filas
contenedorNoticia.config(yscrollcommand=scrollNoticia.set) # Para que el scroll muestre a qué altura se encuentra el cuadro de texto asociado

root.mainloop() # Bucle infinito para escuchar acciones del usuario (siempre tiene que ir al final)
```
