```python
from tkinter import *
```

```python
def CrearVentana():
    nuevoRoot = Tk()
    nuevoRoot.title("Sistemas inteligentes") # Escribir nombre de la ventana
    nuevoRoot.resizable(True, True) # Si se puede escalar a lo ancho y a lo alto respectivamente
    nuevoRoot.iconbitmap("icono256.ico") # Escoger imagen que acompa√±a al titulo
    nuevoRoot.config(bg="#F0F0F0") # Establecer color a la ventana
    return nuevoRoot
```

```python
def CrearFrame(nuevoAncho, nuevoAlto, nuevoColor):
    nuevoFrame = Frame(root, width=nuevoAncho, height=nuevoAlto) 
    nuevoFrame.config(bg=nuevoColor) #Establecer un color al contenedor   
    # Esta linea siempre tiene que ir al final de la funcion
    nuevoFrame.pack() # Reescalar el frame cuando se reescale la ventana
    return nuevoFrame
```

```python
def BuscarNoticias():
    contenedorNoticia.set("Funcion de Aitor")
```

```python
root = CrearVentana()  # Crear frame de la ventana
frameConsulta = CrearFrame("720", "200", "#00FF00") # Crear contenedor para un grupo de elementos (como un div de HTML)
frameConfig = CrearFrame("720", "200", "#0000FF")
frameResultados = CrearFrame("720", "200", "#FF0000")

labelConsulta = Label(frameConsulta, text="Consulta:", bg="#FFFFFF").grid(row=0, column=0, padx=10, pady=10)
cuadroConsulta = Entry(frameConsulta).grid(row=0, column=1, padx=10, pady=10)

labelTopN = Label(frameConfig, text="TOP-N:", bg="#FFFFFF").grid(row=0, column=0, padx=10, pady=10)
desplegableN = Label(frameConfig, text="desplegableN", bg="#FFFFFF").grid(row=0, column=1, padx=10, pady=10)

labelFiltrar = Label(frameConfig, text="Filtrar:", bg="#FFFFFF").grid(row=0, column=2, padx=10, pady=10)
desplegableFiltrar = Label(frameConfig, text="desplegableFiltrar", bg="#FFFFFF").grid(row=0, column=3, padx=10, pady=10)

botonBuscar = Button(frameConfig, text="Buscar", command=BuscarNoticias).grid(row=0, column=4, padx=10, pady=10)

labelRanking = Label(frameResultados, text="Ranking:", bg="#FFFFFF").grid(row=0, column=0, padx=10, pady=10)
scrollRanking = Scrollbar(frameResultados, command=contenedorRanking.yview)
scrollRanking.grid(row=1, column=1, sticky="nsew")
contenedorRanking = Text(frameResultados, width=20, height=8, padx=10, pady=10, bg="#FFFFFF")
contenedorRanking.grid(row=1, column=0)
contenedorRanking.config(yscrollcommand=scrollRanking.set)

labelNoticia = Label(frameResultados, text="Texto de la noticia:", bg="#FFFFFF").grid(row=0, column=2, padx=10, pady=10)
scrollNoticia = Scrollbar(frameResultados, command=contenedorNoticia.yview)
scrollNoticia.grid(row=1, column=3, sticky="nsew")
contenedorNoticia = Text(frameResultados, width=20, height=8, padx=10, pady=10, bg="#FFFFFF")
contenedorNoticia.grid(row=1, column=2)
contenedorNoticia.config(yscrollcommand=scrollNoticia.set)

root.mainloop() # Bucle infinito para escuchar acciones del usuario
```
